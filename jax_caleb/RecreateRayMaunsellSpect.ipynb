```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ray_spect(fs, contrast):
#     spect = (10**3)*np.exp(-fs/5)
    
    a = 1e5
    r = 10
    
    spect = a/(r**2 + fs**2)
    
    if contrast != 0:
        sig = 10
        shift = 4 * np.sqrt(contrast) + 20
        gaussian = 300 * np.exp(-(fs - shift)**2/(np.sqrt(2) * sig)**2)
        spect = spect + gaussian
    
    return spect
```

```python
fs = np.arange(0, 101, 1)
con_range = np.array([0, 25, 50, 100])
ideal_spect = np.zeros((len(fs), len(con_range)))
ind = 0

for rr in con_range:
    ideal_spect[:, ind] = ray_spect(fs, rr)
    ind += 1
    
ideal_spect = ideal_spect/np.mean(ideal_spect)
```

```python
plt.plot(fs, ideal_spect/np.mean(ideal_spect))
```

```python
a = 1e5
r = 10

lzian = a/(r**2 + fs**2)

plt.plot(fs, lzian)
```

```python
sample_fs = np.linspace(0, 100, 30)

target_PS = np.array([np.interp(sample_fs, fs, idl_ps) for idl_ps in ideal_spect.T]).T
```
